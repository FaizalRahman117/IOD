### 3. Prepare the data for clustering

1. Remove the `sex`, `sport` column. We will see if the clusters from K-Means end up like the actual sex.

```python
# ANSWER
#drop columns
ais.drop(columns = ['sex','sport'], inplace = True)
ais.head()
```

### 4. Clustering with K-Means

- Cluster the data to our our target groups.
    - Use sklearn.cluster.KMeans
- We know that there are 2 actual classes. Lets try using 2, 4. Run the clustering algorithm with each of those k values.

```python
# ANSWER
from sklearn.cluster import KMeans
  
kmeans = KMeans(n_clusters = 2) 
kmeans.fit(ais) 
  
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
km2_results = pd.DataFrame(kmeans.labels_)
km2_results[0].value_counts()
```

- Compared to the actual split between males and females, these results are relatively close. (Actual Male = 102, Actual Female = 100)
- Problem is that we can't definitely understand if '0' or '1' refers to male or female. More EDA will be needed to understand this

```python
kmeans = KMeans(n_clusters = 4) 
kmeans.fit(ais) 
  
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

### 5. Get the labels for the models

```python
# ANSWER
kmeans = KMeans(n_clusters = 2) 
kmeans.fit(ais) 
  
print(kmeans.labels_)
```

### 6. Elbow Method For Optimal k

For each k value, we will initialise k-means and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre. As k increases, the sum of squared distance tends to zero. 

```python
# ANSWER
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(ais)
    distortions.append(kmeanModel.inertia_)
```

Below is a plot of sum of squared distances for k in the range specified above. If the plot looks like an arm, then the elbow on the arm is optimal k.

```python
# ANSWER
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
