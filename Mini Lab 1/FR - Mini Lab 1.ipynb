```python
#Import required packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read the various CSVs in the data folder

df90_99 = pd.read_csv('../Mini Lab 1/Raw Data/resale-flat-prices-based-on-approval-date-1990-1999.csv',header = 0)
df00_12 = pd.read_csv('../Mini Lab 1/Raw Data/resale-flat-prices-based-on-approval-date-2000-feb-2012.csv', header = 0)
df12_14 = pd.read_csv('../Mini Lab 1/Raw Data/resale-flat-prices-based-on-registration-date-from-mar-2012-to-dec-2014.csv', header = 0)
df15_16 = pd.read_csv('../Mini Lab 1/Raw Data/resale-flat-prices-based-on-registration-date-from-jan-2015-to-dec-2016.csv', header = 0)
df17_beyond = pd.read_csv('../Mini Lab 1/Raw Data/resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv', header = 0)


```

```python
#two dataframes have 11 columns while everything else has 10
#extra column 'remaining_lease is redundant and can be easily calculated for our own purposes'
#we will drop those columns

df15_16 = df15_16.drop(['remaining_lease'], axis=1)
print(df15_16.shape)
print(df15_16.head())

```

```python
df17_beyond = df17_beyond.drop(['remaining_lease'], axis=1)
print(df17_beyond.shape)
print(df17_beyond.head())
```

```python
df17_beyond.month.unique()
```

```python
help(pd.concat)
```

```python
Resale = pd.concat([df90_99,df00_12, df12_14, df15_16, df17_beyond],axis = 0)
Resale.head()
```

```python
#Confirm that total no.of rows match the sums of the individual csvs
Resale.shape
```

```python
Resale.info()
```

```python
#append a day '-01' to the end of the month category in order to have it read like a full date
#this is an assumption of course, but is alright since we will not be doing any analysis on the actual transaction day

Resale['month'] = Resale['month'] + str('-01') 
```

```python
Resale
```

```python
#check null values
Resale.isnull().sum()
```

```python
#understand the various categorical variables of the dataset

print(Resale['town'].unique())
print(Resale['storey_range'].unique())
print(Resale['flat_model'].unique())
print(Resale['flat_type'].unique())
```

```python
#you will realize that the 'flat_model' column has some duplicate values that are fully caps
#We can fix this by converting all strings to titles (first letter of every word is caps)
#Flat type also has duplicates for the multi generation category, we can use replace

#Resale = Resale.replace({'flat_model' : { 'IMPROVED' : 'Improved', 'NEW GENERATION' : 'New Generation', 
#                                         'MODEL A' : 'Model A', 'STANDARD' : 'Standard', 'SIMPLIFIED' : 'Simplified', 
#                                         'MODEL A-MAISONETTE' : 'Model A-Maisonette', 'APARTMENT' : 'Apartment', 
#                                        'MAISONETTE' : 'Maisonette', 'TERRACE' : 'Terrace', '2-ROOM' : '2-room',
#                                      'IMPROVED-MAISONETTE' :'Improved-Maisonette', 'MULTI GENERATION' : 'Multi Generation',
#                                       'PREMIUM APARTMENT' : 'Premium Apartment' }})



Resale['flat_model'] = Resale['flat_model'].str.title() 

Resale['flat_model'].unique()
```

```python
Resale = Resale.replace({'flat_type' : {'MULTI GENERATION' : 'MULTI-GENERATION'}}) 
Resale['flat_type'] = Resale['flat_type'].str.title()
Resale['flat_type'].unique()
```

```python
#Let's also convert town name & flat types to titles in order for it to read better

Resale['town'] = Resale['town'].str.title()

print(Resale['town'].unique())
```

```python
#understand the numerical part of the dataframe
Resale.describe()
```

```python
#we can also calculate the price per sqm based on the current columns
Resale['price_per_sqm'] = Resale['resale_price'] / Resale['floor_area_sqm']
```

```python
#look at the finished cleaned product
Resale
```

```python
#Export CSV File
Resale.to_csv('HDB_Resale_Complete.csv', index=False)
```
