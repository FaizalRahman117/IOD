```
# print the list of students
# in the following format
# Index 01 	 Alice
# Index 02 	 Bob
# Index 03 	 Pradeep

#Define students
students = ["frank", "sinatra", "jake", "steve", "paul", "james", "stevie", "wonder", "juke", "jameson"]

#for loop
for i in range(10):
    print("Index 0"+str(i+1)+" "+str(students[i]))
```

```
# print the list of students in reverse order one student in a line with its index
#Define students


#for loop
for i in range(10,0,-1):
    print("Index 0"+str(i)+" "+str(students[i-1]))
```

```
# print the student list in random order
# hint look up the function random.sample
import random as random
students_random = random.sample(students,10)
print (students_random)
```

```
# list students in order of their exam marks in the following list
students = ["frank", "sinatra", "jake", "steve", "paul", "james", "stevie", "wonder", "juke", "jameson"]
marks = [87,45,76,64,66,90,55,73,69,75]
studentsAndMarks = dict(zip(students,marks))
std_list = list(studentsAndMarks.items())
print(sorted(std_list,key = lambda x : x[1],reverse = True))
```
